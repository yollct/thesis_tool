```python
clusters = pd.read_csv("/home/chit/Desktop/Thesis/results/5.08/cluster_table.csv", sep="\t")
clusters
```

```python
with open("/home/chit/Desktop/Thesis/results/5.08/all_connect.txt") as f:
    connect = f.read().split("\n")
```

```python
logodd_cluster = []
for a in connect:
    split = a.split("_")
    for b in split:
        c = "Cluster "+ b
        if c not in logodd_cluster:
            logodd_cluster.append(c)
```

```python
##filter the clusters and oxi gene list
highlogodd = clusters[clusters['cluster'].isin(logodd_cluster)].dropna()
gene_list = highlogodd[highlogodd['object'].isin(oxi_gene['Gen'])].ENSEMBL.to_list()
gene_list
```
